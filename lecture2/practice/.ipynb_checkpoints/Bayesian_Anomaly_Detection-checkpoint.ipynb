## Bayesian modeling with STAN

Modelamos que la cantidad de consultas que hace la clÃ­nica $n$ el dÃ­a $t$ es sampleado de una Poisson cuyo rate es el producto del tamaÃ±o de la clÃ­nica $s_n$ por un indicador de la magnitud del dÃ­a, $d_t$: consultas ~ Poissson($s_n\, d_t$).  

```python
Image("graphical_model.png")
```

```python
# definimos el modelo Bayesiano para ejecutar con STAN
# (Observar que no es de la clase de modelos que tiene Likelihood, simplemente el modelo es igualar los datos a una funciÃ³n de los parÃ¡metros)
my_model="""
data {
 int <lower=0> N; // number of clinics
 int <lower=0> T; // number of days in the dataset
 int <lower=0> my_max; // max # of practices in the dataset to break degeneracies
 array[T,N] int<lower=0> practices;  // practices on day t in clinic n: practices[t,n]
 
}

parameters {
  vector<lower=0, upper=1>[T] d; // parameter that measures how 'hot' is each day
  vector<lower=0, upper=my_max>[N] s; // parameter that estimates the size of each clinic
  real <lower=0> eps;
}

model {

   for (n in 1:N) {
       for (t in 1:T) {
           eps ~ normal(0, d[t]*s[n]/10 ); // A clinic can be randomly smaller because some missing doctors
           practices[t,n] ~ poisson( d[t]*s[n] - eps );
     }; 
     };
}
"""
```

```python
int(data[list(data)[1:]].max().max())
```

```python
# Definimos el diccionario que contiene la data que necesita el modelo
# Observar que my_max es la cantidad mÃ¡xima diaria de consultas del dataset.  El modelo lo hicimos de modo que s*d en ese 
# caso requiera que d = 1, ya que s se samplea hasta my_max.  Esto es necesario para que no haya degeneraciÃ³n.  Si no 
# habrÃ­a diversas soluciones todas igual de vÃ¡lidas, ya que uno puede bajar d y aumentar s, y sigue siendo todo igual de vÃ¡lido
# ParecerÃ­a ser que funcionÃ³ este truco
my_data = {'N': len(list(data)), 'T':len(data), 'my_max': int(data[list(data)[1:]].max().max()), 'consultas': np.array(data[list(data)[1:]]).astype(int) }
```

```python
# Metemos la data al modelo para construir el modelo
model = stan.build(my_model, data=my_data, random_seed=1);
```

```python
# Lo fiteamos.  
# o sea, sampleamos de la posterior para toas las variables del modelo
fit = model.sample(num_chains=4, num_samples=1000, delta=080)
```

```python
# Construimos el dataframe de la posterior ya que es lo que necesitamos para hacer nuestro trabajo
df = fit.to_frame() 
df
```

```python
# Verificamos que el parÃ¡metro rhat < 1.05, lo que nos da la certeza de que el sampleo estÃ¡ en equilibrio e insesgado
azdata = az.from_pystan(fit)
summary = az.summary(azdata)
summary
```

The rhat should be below to 1.01 (1.05 acceptable) to have a fully confident sampling.  It is "ok"

```python
plt.hist(summary['r_hat'])
plt.show()
```

### Let's sort the clinics

We sort them from most unlikely because being above the expectation, to the most unlikey for being below the expectation.  All this corresponds to the given day of study.

We can start by computing the average p-value 
$$\langle p_n \rangle = \langle \sum_{x=consultas[n]@d_0}^\infty PoissonPDF(x; d_0 \cdot s[n]) \rangle $$
for every clinic $n$, where $d_0$  is the day of study.  The clinic that has the minimum $\langle p_n \rangle$ is the most unlikely from above, that we're looking for.

```python
# Dataframe with all the required info to compute <p_n>
my_day = df[['d.'+str(len(data3))]+['s.'+str(i+1) for i in range(len(list(data3)))]]
my_day[:4]
```

```python
clinics = list(data3)
probabilities = pd.DataFrame()
for i in range(len(clinics)):
    if i % 10 ==0: print(i,'/',len(clinics))
    tmp = []
    for k in range(len(my_day)):
    #for k in range(10):
        #k = np.random.randint(0,len(lastday))
        # Here I compute a sum to infinity using the CDF of the Poisson distribution
        tmp.append( 1 - stats.poisson.cdf(data3.iloc[-1][clinics[i]], my_day.iloc[k]['d.'+str(len(data3))] * my_day.iloc[k]['s.'+str(i+1)]))
    probabilities[clinics[i]] = [np.mean(tmp), np.std(tmp),int(data3.iloc[-1][clinics[i]]), np.mean(my_day['d.'+str(len(data3))]) * np.mean(my_day['s.'+str(i+1)]), [my_day.iloc[k]['d.'+str(len(data3))] * my_day.iloc[k]['s.'+str(i+1)] for k in range(len(my_day))]]
probabilities.index = ['<p_n>','SD(p_n)','consultas_my_day','Poisson_mean_rate', 'rate_list']
```

```python
# This new dataframe has all the info!
# Lo ordenamos de menor a mayor en <p_n>, de modo que la primer clÃ­nica es aquella que 
# tuvo consultas mÃ¡s por arriba de lo esperado
probabilities.sort_values(by = '<p_n>', axis = 1, inplace=True) 
probabilities
```
