### Playground

```python
xmin= 0
xmax = 14
np.random.seed(1)
x = np.linspace(xmin, xmax, 100)
plt.scatter([7,4,6,5,9,5.4, 6.1],[0,0,0,0,0,0,0], marker = 'x',  color='red', label='Data')
for i in range(60):
    plt.plot(x, stats.norm.pdf(x, np.random.normal(6.4,1),np.random.normal(2,0.4)), 'b-', alpha=0.3)
plt.ylim(-0.01,0.3)
plt.xlim(xmin, xmax)
plt.yticks([])
plt.legend()
plt.show()
```

```python
data = []
for i in range(5000):
    data.append(stats.norm.rvs(np.random.normal(6.4,1),np.random.normal(2,0.4)))
```

```python
xmin= 0
xmax = 14
np.random.seed(1)
x = np.linspace(xmin, xmax, 100)
plt.scatter([7,4,6,5,9,5.4, 6.1],[0,0,0,0,0,0,0], marker = 'x',  color='red', label='Data')
plt.hist(data, histtype='step', bins=20, density=1, color='blue', label='What the model expects')
plt.ylim(-0.01,0.3)
plt.xlim(xmin, xmax)
plt.text(1,0.25,'Sample tons of data', fontsize=12)
plt.yticks([])
plt.legend()
plt.show()
```
