## Crypto price as Gaussian Mixture

```python
sigmas = [10,100,1000]
crypto_price = [50000]
np.random.seed(1)
z = np.random.choice([0,1,2], p=[0.7,0.25,0.05], size=5000)
for zn in z:
    crypto_price.append(cripto_price[-1]+np.random.normal(0,sigmas[zn]))
```

```python
plt.plot(cripto_price)
plt.title('Emulated crypto price')
plt.ylabel('USD')
plt.xlabel('Time')
plt.show()
```
